# List of Benchmark Functions

- List of benchmark function examples [see this link](https://www.sfu.ca/~ssurjano/optimization.html)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def grapher(X, Y, Z):
    fig = plt.figure(figsize=(6, 4), dpi=100)
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z)
    plt.show()
```

## Sphere function

formula: ${f(x_{1} \cdots x_{n})=\sum_{i=1}^{n}x_{i}^2}$

where:
${
\infty \leqq x_{i} \infty \\
f_{min}(0, \cdots , 0)=0
}$

```python
def sphere(x, y):
    t1 = 0.1 * x**2
    t2 = y**2
    return t1 + t2

x = np.arange(-5.0, 5.0, 0.1)
y = np.arange(-5.0, 5.0, 0.1)

X, Y = np.meshgrid(x, y)
Z = sphere(X, Y)
# print('X', X.shape, 'Y', Y.shape, 'Z', Z, sep="\n")

grapher(X, Y, Z)
```

## Ackley function

formula: ${f(x_{1} \cdots x_{n})=20-20\exp \biggl( -0.2\sqrt{\frac{1}{n}\sum_{i=1}^{n}x_{i}^2} \biggr) +e-\exp \biggl(\frac{1}{n}\sum_{i=1}^{n}\cos(2\pi x_{i}) \biggr)}$

where:
$
{-32.768 \leqq x_{i} \leqq 32.768 \\
f_{min}(0, \cdots , 0)=0
}
$

```python
def ackley(x, y):
    t1 = 20
    t2 = -20 * np.exp(-0.2 * np.sqrt(0.5 * (x**2 + y**2)))
    t3 = np.e
    t4 = -np.exp(0.5 * (np.cos(2 * np.pi * x) + np.cos(2 * np.pi * y)))
    return t1 + t2 + t3 + t4

x = np.arange(-30, 30, 0.1)
y = np.arange(-30, 30, 0.1)

X, Y = np.meshgrid(x, y)
Z = ackley(X, Y)

grapher(X, Y, Z)
```

## Rosenbrock

formula: ${f(x_{1} \cdots x_{n-1})=\sum_{i=1}^{n}(100(x_{i+1}-x_{i}^2)^2+(x_{i}-1)^2)}$

where:
$
{-5 \leqq x_{i} \leqq 5 \\
f_{min}(1, \cdots , 1)=0
}
$

```python
def rosenbrock(x, y):
    return 0.5 * (100 * (y - x**2)**2 + (x - 1)**2)

x = np.arange(-5, 5, 0.1)
y = np.arange(-5, 5, 0.1)

X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

grapher(X, Y, Z)
```

## Bukin function

formula: ${f(x_{1},x_{2})=100*\sqrt{|x_{2}-0.01x_{1}^2|}+0.01|x_{1}+10|}$

where:
$
−15≦x1≦−5−3≦x2≦3fmin(−10,1)=0 
$

```python
def bukin(x, y):
    return 100 * np.sqrt(np.abs(y - 0.01 * x)) + 0.01 * np.abs(x + 10)

x = np.arange(-30, 30, 0.1)
y = np.arange(-30, 30, 0.1)

X, Y = np.meshgrid(x, y)
Z = bukin(X, Y)

grapher(X, Y, Z)
```

## Levi function

formula: ${f(x_{1},x_{2})=\sin^2(3\pi x_{1})+(x_{1}-1)^2\bigl(1+\sin^2(3\pi x_{2})\bigr)+(x_{2}-1)^2\bigl(1+\sin^2(2\pi x_{2})\bigr)}$

where:
$
−10≦x1,x2≦10 \\
fmin(1,1)=0
$

```python
def levi(x, y):
    t1 = np.sin(3 * np.pi * x)**2
    t2 = (x - 1)**2 * (1 + np.sin(3 * np.pi * y)**2)
    t3 = (y - 1)**2 * (1 + np.sin(2 * np.pi * y)**2)

    return t1 + t2 + t3

x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(x, y)
Z = levi(X, Y)

grapher(X, Y, Z)
```

## Eggholder function

formula: ${f(x_{1},x_{2})=-(x_{2}+47)\sin\bigl(\sqrt{|x_{2}+\frac{x_{1}}{2}+47|}\bigr)-x_{1}\sin\bigl(\sqrt{|x_{1}-(x_{2}+47)|}\bigr)}$

where:
$
{-512 \leqq x_{1} , x_{2} \leqq 512 \\
f_{min}(512,404.2319)=-959.6407
}
$

(a bit too broad range)

```python
def eggholder(x, y):
    t1 = -(y + 47) * np.sin(np.sqrt(np.abs(y + 0.5 * x + 47)))
    t2 = x * np.sin(np.sqrt(np.abs(x - (y + 47))))

    return t1 + t2

x = np.arange(-500, 500, 0.5)
y = np.arange(-500, 500, 0.5)

X, Y = np.meshgrid(x, y)
Z = eggholder(X, Y)

grapher(X, Y, Z)
```

## Rastrigin function

formula: ${f(x_{1} \cdots x_{n})=10n+\sum_{i=1}^{n}(x_{i}^2-10\cos(2 \pi x_{i}))}$

where:
$
{-5.12 \leqq x_{i} \leqq 5.12 \\
f_{min}(0, \cdots , 0) = 0
}
$

```python
def rastrigin(x, y):
    t1 = 20 + x**2 - 10 * np.cos(2 * np.pi * x)
    t2 = y**2 - 10 * np.cos(2 * np.pi * y)

    return t1 + t2

x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(x, y)
Z = rastrigin(X, Y)

grapher(X, Y, Z)
```

## Six-hump camel function

formula: ${f(x_{1},x_{2}) = (4-2.1x_{1}^2+\frac{1}{3}x_{1}^4)x_{1}^2+x_{1}x_{2}+4(x_{2}^2-1)x_{2}^2}$

where:
$
{-3 \leqq x_{1} \leqq 3 \\
-2 \leqq x_{2} \leqq 2 \\
f_{min}(0.0898,-0.7126) = f_{min}(-0.0898,0.7126) \approx -1.0316
}
$

```python
def sixhump_camel(x, y):
    t1 = (4 - 2.1 * x**2 + (1 / 3) * x**4) * x**2
    t2 = x * y
    t3 = 4 * (y**2 - 1) * y**2

    return t1 + t2 + t3

x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(x, y)
Z = sixhump_camel(X, Y)

grapher(X, Y, Z)
```

## De Jong's function 3

formula: {f(x_{1} \cdots x_{5})=\sum_{i=1}^{5}[x_{i}]}

where:
$
{-5.12 \leqq x_{1} \leqq 5.12 \\
-5.12 \leqq x_{2} \leqq 5.12 \\
f_{min}(-5.12 \sim -5, \cdots , -5.12 \sim -5)=0
}
$

```python
def de_jong(x, y):
    t1 = np.floor(x)
    t2 = np.floor(y)

    return t1 + t2

x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(x, y)
Z = de_jong(X, Y)

grapher(X, Y, Z)
```

## Styblinski-Tang function

formula: ${f(x_{1} \cdots x_{n})=\frac{\sum_{i=1}^{n}x_{i}^4-16x_{i}^2+5x_{i}}{2}}$

where:
$
{-5 \leqq x_{1} \leqq 4 \\
-3 \leqq x_{2} \leqq 4 \\
-39.16617n \leq f(-2.903534, \cdots , -2.903534) \leq -39.16616n \\
f_{min}(-2.903534, \cdots , -2.903534) \approx -39.166165n
}
$

```python
def styblinski_tang(x, y):
    def temp(t):
        return t**4 - 16 * t**2 + 5 * t

    return (temp(x) + temp(y)) * 0.5

x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(x, y)
Z = styblinski_tang(X, Y)

grapher(X, Y, Z)
```

## Matyas function

formula: ${f(x_{1},x_{2})=0.26(x_{1}^2+x_{2}^2)-0.48x_{1}x_{2}}$

where:
$
{-10 \leqq x_{1} , x_{2} \leqq 10 \\
f_{min}(0,0)=0
}
$

```python
def matyas(x, y):
    t1 = 0.26 * (x**2 + y**2)
    t2 = 0.48 * x * y
    
    return t1 + t2

x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(x, y)
Z = matyas(X, Y)

grapher(X, Y, Z)
```

## Goldstein-Price function

formula: ${\begin{split}
f(x_{1},x_{2})= & \Bigl(1+(x_{1}+x_{2}+1)^2(19-14x_{1}+3x_{1}^2-14x_{2}+6x_{1}x_{2}+3x_{2}^2)\Bigr) \\
& \Bigl(30+(2x_{1}-3x_{2})^2(18-32x_{1}+12x_{1}^2+48x_{2}-36x_{1}x_{2}+27x_{2}^2)\Bigr)
\end{split}}$

where:
$
{-2 \leqq x_{1} , x_{2} \leqq 2 \\
f_{min}(0,-1)=3
}
$

```python
def goldstein_price(x, y):
    t1 = 1 + (x + y + 1)**2 * (19 - 14 * x + 3 * x**2 - 14 * y + 6 * x * y + 3 * y**2)
    t2 = 30 + (2 * x - 3 * y)**2 * (18 - 32 * x + 12 * x + 48 * y - 36 * x * y + 27 * y**2)

    return t1 * t2

x = np.arange(-10, 10, 0.5)
y = np.arange(-10, 10, 0.5)

X, Y = np.meshgrid(x, y)
Z = goldstein_price(X, Y)

grapher(X, Y, Z)
```
